```python
import numpy as np
import pandas as pd

# 1. Create a NumPy array from a list
list_data = [1, 2, 3, 4, 5]
numpy_array = np.array(list_data)
print('NumPy Array from List:')
print(numpy_array)
```

```python
# 2. Generate an array of zeros, ones, or a specific constant in NumPy
zeros_array = np.zeros((3, 3))
ones_array = np.ones((2, 4))
constant_array = np.full((2, 2), 7)

print('Array of Zeros:')
print(zeros_array)
print('Array of Ones:')
print(ones_array)
print('Array of Specific Constant (7):')
print(constant_array)
```

```python
# 3. Find the maximum, minimum, and mean of a NumPy array
array = np.array([1, 5, 7, 3, 9, 2])
max_value = np.max(array)
min_value = np.min(array)
mean_value = np.mean(array)

print('Maximum Value:')
print(max_value)
print('Minimum Value:')
print(min_value)
print('Mean Value:')
print(mean_value)
```

```python
# 4. Stack two NumPy arrays vertically and horizontally
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

vertical_stack = np.vstack((array1, array2))
horizontal_stack = np.hstack((array1, array2))

print('Vertical Stack:')
print(vertical_stack)
print('Horizontal Stack:')
print(horizontal_stack)
```

```python
# 5. Find unique elements in a NumPy array
array_with_duplicates = np.array([1, 2, 2, 3, 4, 4, 5])
unique_elements = np.unique(array_with_duplicates)

print('Unique Elements:')
print(unique_elements)
```

```python
# 6. Generate a random number matrix of size 5x5 using NumPy
random_matrix = np.random.rand(5, 5)

print('Random Matrix (5x5):')
print(random_matrix)
```

```python
# 7. Extract an element in the matrix
extracted_element = random_matrix[2, 3]  # element at row 2, column 3

print('Extracted Element (row 2, column 3):')
print(extracted_element)
```

```python
# 8. Create a Pandas DataFrame from a dictionary
data_dict = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df_dict = pd.DataFrame(data_dict)

print('Pandas DataFrame from Dictionary:')
print(df_dict)
```

```python
# 9. Merge or concatenate two Pandas DataFrames
data_dict2 = {'Name': ['David', 'Eva'], 'Age': [40, 45]}
df_dict2 = pd.DataFrame(data_dict2)

concatenated_df = pd.concat([df_dict, df_dict2], ignore_index=True)

print('Concatenated DataFrame:')
print(concatenated_df)
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = r"D:\Session 2\country_wise_latest_modified.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
#10.Fill missing values
df.fillna(df.mean(numeric_only=True), inplace=True)
df.isnull().sum()
```

```python
# 11.Drop a column or a row
df_dropped_col = df.drop(columns=['WHO Region'], errors='ignore')
df_dropped_row = df.drop(index=0)
df_dropped_col.head()
```

```python
# 12 Export a Pandas DataFrame to a CSV or Excel file
df.to_csv("D:/Session 2/covid_data_cleaned.csv", index=False)
df.to_excel("D:/Session 2/covid_data_cleaned.xlsx", index=False)
print("Files saved successfully.")
```

```python
# 13.Filter rows based on a condition
high_cases = df[df['Confirmed'] > 100000]
high_cases.head()
```

```python
# 14 Scatter plot - Relationship between confirmed cases and deaths
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Confirmed'], y=df['Deaths'])
plt.xlabel("Confirmed Cases")
plt.ylabel("Deaths")
plt.title("Confirmed Cases vs Deaths")
plt.show()
```

```python
# 15.Box plot - Variation of active cases across different countries
plt.figure(figsize=(12, 6))
sns.boxplot(y=df['Active'])
plt.ylabel("Active Cases")
plt.title("Variation of Active Cases Across Different Countries")
plt.show()
```

```python
# 16 Bar chart - New cases comparison across WHO regions
plt.figure(figsize=(10, 5))
df.groupby('WHO Region')['New cases'].sum().plot(kind='bar', color='skyblue')
plt.xlabel("WHO Region")
plt.ylabel("New Cases")
plt.title("New Cases Comparison Across WHO Regions")
plt.xticks(rotation=45)
plt.show()
```

```python
# 17.Line plot - Change in confirmed cases compared to last week
plt.figure(figsize=(10, 5))
df.groupby('Country/Region')['Confirmed last week'].sum().plot(kind='line', marker='o', color='green')
plt.xlabel("Country")
plt.ylabel("Confirmed Cases Last Week")
plt.title("Confirmed Cases Compared to Last Week")
plt.xticks(rotation=90)
plt.grid()
plt.show()

```
